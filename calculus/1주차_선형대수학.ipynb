### 벡터의 정의와 연산 
scalar는 크기만 가지는 양이다. vector는 크기와 방향을 가지는 양이다. 

이를 표현하기 위해서 우리는 아래와 같이 그 방향의 존재를 표기하기로 하였다. 
$$ \overrightarrow{AB} $$ 

벡터 $\overrightarrow{AB}$ 의 크기는  $ \left | \overrightarrow{AB} \right | $ 로 표현한다. 

$$
(a1,a2) \in \mathbb{R^2}, (a1,a2,a3) \in \mathbb{R^3},(x_1,x_2,x_3,x_4, \dots, x_n) \in \mathbb{R^n} \\
\vec{a} = (a_1,a_2) , then \left | \vec{a} \right |  = \sqrt{a_1^2 + a_2^2} \\
\vec{a} =(a_1,a_2,a_3) then \left | \vec{a} \right | = \sqrt{a_1^2 + a_2^2 + a_3^2} 
$$

위와 같이 우리는 벡터를 정의하였다. 

자 그럼 이제 벡터의 연산을 배워볼 차례이다. 
임의의 벡터 $\vec{a}=(a_1,a_2)$, $\vec{b}=(b_1,b_2)$

 이것은 영벡터이다.

### 벡터의 내적 

일반적으로 아는 형태의 내적이다. 내적의 값은 스칼라이다. 
이 때 
$$ 
\vec{a},\vec{b} \in \mathbb{R^2}\\
\vec{a}\cdot\vec{b} = (a_1,a_2)\cdot(b_1,b_2) =a_1b_1 + a_2b_2
$$ 
위의 과정의 일반화가 된다면 벡터의 내적은 아래와 같이 정의된다. 
$$
\vec{a},\vec{b} \in \mathbb{R^n}\\
\vec{a}\cdot\vec{b} = (a_1,a_2,\dots,a_n)\cdot(b_1,b_2,\dots,b_n) =a_1b_1 + a_2b_2 + \dots + a_n b_n = \sum_{k=1}^{k} a_kb_k 
$$
cross product 
$$ \vec{p}\times \vec{q}=|\vec{p}|\vec{q}|sin\theta \hat{n} $$

dot product 
$$ \vec{p}\cdot\vec{q}=|\vec{p}|\vec{q}|cos\theta $$

따라 우리는 내적을 이와 같이 나타낼 수도 있다. 
$$
cos\theta  = \frac{\vec{p}\cdot\vec{q}}{|\vec{p}|\vec{q}|}
$$ 

따라 우리는 여기서 영벡터가 아닌 두벡터에 대해 
$$ 
\vec{a}\cdot\vec{b} = 0 \Leftrightarrow \vec{a}\perp \vec{b}
$$
